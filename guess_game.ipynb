```python
import random 
number=random.randint(70,90)
print("you guess the number between 70 to 90")
attempts=7
print("you have got 7 attempts")
for i in range(attempts):
    guess=int(input("enter the number"))
    if number == guess:
        print(" you guessed the number correct")
        break
    if number < guess:
        print(" your guess is greater")
    if number > guess:
        print(" your guess is lower")

if guess!= number:
    print("your out of attempts")
    print(" the correct number is", number)


        
              


```
