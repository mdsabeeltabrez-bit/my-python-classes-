```python
try:
    num1=int(input("enter the number1 "))
    num2=int(input("enter the number2 "))
    result=num1/num2
    print(result)

except ValueError:
    print("enter number instead of alphabet")

except ZeroDivisionError:
    print(" a number cannot be divided by zero")

except SyntaxError:
    print("there is issue with spelling ")

else:
    print("no exception")

finally:
    print("it marks the end of code")
```
